```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Leaky Rectified Linear Unit (leaky ReLU) Activation Function


def leaky_ReLU(x, slope):
    data = [max(slope*value, value) for value in x]
    return np.array(data, dtype=float)



# Derivative for leaky ReLU 
def der_leaky_ReLU(x, slope):
  data = [1 if value>0 else slope for value in x]
  return np.array(data, dtype=float)

# Generating data For Graph
x_data = np.linspace(-10,10,100)
y_data = leaky_ReLU(x_data, .01)
dy_data = der_leaky_ReLU(x_data, .01)

# Graph
plt.plot(x_data, y_data, x_data, dy_data)
plt.title('leaky ReLU Activation Function & Derivative')
plt.legend(['leaky_ReLU (alpha=0.01)','der_leaky_ReLU (alpha=0.01)'])
plt.grid()
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig("leaky_relu.jpg")
plt.show()
```

```python
def ReLU(x):
    data = [max(0, value) for value in x]
    return np.array(data, dtype=float)

def der_ReLU(x):
  data = [1 if value>0 else 0 for value in x]
  return np.array(data, dtype=float)

# Generating data For Graph
x_data = np.linspace(-5,5,100)
y_data = ReLU(x_data)
dy_data = der_ReLU(x_data)

# Graph
plt.plot(x_data, y_data, x_data, dy_data)
plt.title(' ReLU Activation Function & Derivative')
plt.legend(['ReLU','der_ReLU'])
plt.grid()
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig("relu.jpg")
plt.show()
```

```python
def softplus(x, beta):
    # Avoid overflow of log(1+exp(B*x)) = log(1+exp(B*x)) - log(exp(B*x)) + B*x = log(1+exp(-B*x)) + B*x
    data = [(math.log(1+math.exp(-abs(value*beta))) +
             max(value*beta, 0))/beta for value in x]
    return np.array(data, dtype=float)

def derv_softplus(x, beta):
    # log(1+exp(-B*x)) + B*x -> dy/dx = (-B*exp(-B*x))/(1+exp(-B*x)) + B
    data = [-math.exp(-(value*beta))/(1+math.exp(-(value*beta))) + 1
              for value in x]
    return np.array(data, dtype=float)

    # Generating data For Graph
x_data = np.linspace(-5, 5, 100)
y_data = softplus(x_data, 1)
dy_data = derv_softplus(x_data, 1)

# Graph
plt.plot(x_data, y_data, x_data, dy_data)
plt.title('Softplus Activation Function and Derivative')
plt.legend(['Softplus (beta=1)', 'derv_Softplus (beta=1)'])
plt.grid()
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig("softplus.jpg")
plt.show()
```

```python
from math import erf
def GELU(x):
    data = [0.5 * value * (1.0 + erf(value / np.sqrt(2.0))) for value in x]
    return np.array(data, dtype=float)

def derv_GELU(x):
    data = [0.5*(1+erf(value / np.sqrt(2.0)))+value*math.exp((-value**2)/2)/np.sqrt(2*np.pi) for value in x]
    return np.array(data, dtype=float)

    # Generating data For Graph
x_data = np.linspace(-5, 5, 100)
y_data = GELU(x_data)
dy_data = derv_GELU(x_data)
#dy_data = der_leaky_ReLU(x_data)

# Graph
plt.plot(x_data, y_data, x_data, dy_data)
plt.title('GELU Activation Function and Derivative')
plt.legend(['GELU', 'GELU Derivative'])
plt.grid()
plt.xlabel('Input')
plt.ylabel('Output')
plt.savefig("gelu.jpg")
plt.show()
```
