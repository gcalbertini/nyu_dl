```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Leaky Rectified Linear Unit (leaky ReLU) Activation Function


def leaky_ReLU(x, slope):
    data = [max(slope*value, value) for value in x]
    return np.array(data, dtype=float)

# Derivative for leaky ReLU
# def der_leaky_ReLU(x):
#  data = [1 if value>0 else 0.05 for value in x]
#  return np.array(data, dtype=float)


# Generating data For Graph
x_data = np.linspace(-5, 5, 100)
y_data = leaky_ReLU(x_data, 0.01)
#dy_data = der_leaky_ReLU(x_data)

# Graph
plt.plot(x_data, y_data)
plt.title('leaky ReLU Activation Function')
plt.legend(['leaky_ReLU (neg_slope = 0.01)'])
plt.grid()
plt.savefig("leaky_relu.jpg")
plt.show()
```

```python
def ReLU(x):
    data = [max(0, value) for value in x]
    return np.array(data, dtype=float)

    # Generating data For Graph
x_data = np.linspace(-5, 5, 100)
y_data = ReLU(x_data)
#dy_data = der_leaky_ReLU(x_data)

# Graph
plt.plot(x_data, y_data)
plt.title('ReLU Activation Function')
plt.legend(['ReLU'])
plt.grid()
plt.savefig("relu.jpg")
plt.show()
```

```python
def softplus(x, beta):
    # log(1+exp(B*x)) = log(1+exp(B*x)) - log(exp(B*x)) + B*x = log(1+exp(-B*x)) + B*x
    data = [(math.log(1+math.exp(-abs(value*beta))) +
             max(value*beta, 0))/beta for value in x]
    return np.array(data, dtype=float)


    # Generating data For Graph
x_data = np.linspace(-5, 5, 100)
y_data = softplus(x_data, 1)
#dy_data = der_leaky_ReLU(x_data)

# Graph
plt.plot(x_data, y_data)
plt.title('Softplus Activation Function')
plt.legend(['Softplus (beta=1)'])
plt.grid()
plt.savefig("softplus.jpg")
plt.show()
```

```python
from math import erf


def GELU(x):
    data = [0.5 * value * (1.0 + erf(value / np.sqrt(2.0))) for value in x]
    return np.array(data, dtype=float)

    # Generating data For Graph
x_data = np.linspace(-5, 5, 100)
y_data = GELU(x_data)
#dy_data = der_leaky_ReLU(x_data)

# Graph
plt.plot(x_data, y_data)
plt.title('GELU Activation Function')
plt.legend(['GELU'])
plt.grid()
plt.savefig("gelu.jpg")
plt.show()
```
